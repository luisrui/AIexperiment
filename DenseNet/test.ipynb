```python
import pandas as pd
data = pd.read_csv('./data/all/images.txt')
list(data['img_path'])
```

```python
labels = pd.read_csv('./data/all/labels.txt')
```

```python
train_imgs, test_imgs, train_labels, test_labels = ['img_path'], ['img_path'], ['labels'], ['labels']
for i, path in enumerate(list(data['img_path'])):
    if i % 15 == 3 or i % 15 == 7:
        test_imgs.append(path)
        test_labels.append(list(labels['labels'])[i])
    else:
        train_imgs.append(path)
        train_labels.append(list(labels['labels'])[i])
```

```python
with open('train_imgs.txt', 'w') as f:
    for line in train_imgs:
        f.write(line + '\n')

with open('test_imgs.txt', 'w') as f:
    for line in test_imgs:
        f.write(line + '\n')

with open('train_labels.txt', 'w') as f:
    for line in train_labels:
        f.write(line + '\n')

with open('test_labels.txt', 'w') as f:
    for line in test_labels:
        f.write(line + '\n')
```
